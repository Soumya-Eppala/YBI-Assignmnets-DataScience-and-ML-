This transformer should be used to encode target values i.e. y

```python
y=df['Placement']
X=df.drop(['Placement'],axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
le.fit(y)
```

```python
le.transform(y)
```

```python
le.fit_transform(y)           # this helps to fit the tardet value and also transforms to numerical data
```
